```python
import re
import numpy as np
```

```python
action = []
actionStatus = []
numbers = re.compile('-?\d+')
file = open("/home/marmot/catkin_ws/src/marmot/database/rawData.txt", "r")
for i in file:
    temp = list()
    if(i.find('status')!=-1):
        temp = i[39:]
        result = list(map(int, numbers.findall(temp)))
        actionStatus.append(result)
    elif(i.find('actions')!=-1):
        temp = i[24:]
        result = list(map(int, numbers.findall(temp)))
        action.append(result)

print(action)
print(actionStatus)   
```

```python
action = np.array(action)
actionStatus = np.array(actionStatus)
```

```python
assert(action.shape == actionStatus.shape)
```

```python
actionPlan = np.zeros_like(action)

for i in range(action.shape[0]):
    for j in range(action.shape[1]):
        if(actionStatus[i][j]>=0):
            actionPlan[i][j] = action[i][j]
```
